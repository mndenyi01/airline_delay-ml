```python
# Loading necessary libraries and the dataset
import pandas as pd

df = pd.read_parquet("../data/processed/flights_cleaned_base.parquet")
df.shape
```

```python
# Extracting datetime features
df["year"] = df["fl_date"].dt.year
df["month"] = df["fl_date"].dt.month
df["day"] = df["fl_date"].dt.day
df["day_of_week"] = df["fl_date"].dt.dayofweek # Monday=0, Sunday=6
df["is_weekend"] = (df["day_of_week"] >= 5).astype(int)  # 1 if weekend, 0 if weekday
```

Date features insights and all insights in general
